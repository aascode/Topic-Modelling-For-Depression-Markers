```
import pandas as pd
import time
```

```
from google.colab import files
uploaded = files.upload()
```

```
import io
pre = pd.read_excel(io.BytesIO(uploaded['All_tweets_LIWC_results.xlsx']), sheet_name='pre_tweets')
post = pd.read_excel(io.BytesIO(uploaded['All_tweets_LIWC_results.xlsx']), sheet_name='post_tweets')
cg = pd.read_excel(io.BytesIO(uploaded['ControlGroup_allTweets_230_LIWC.xlsx']))
```

```
cg.sort_values(by=['CG_ID'], ignore_index=True, inplace=True)
cg.head()
```

```
user =[]
for i in range(len(cg.columns)):
  user.append(cg.columns[i])
#print(user)

for i in range(len(user)):
  user[i] = user[i] + "_cg"
#print(user)

cg.columns = user
cg.head()
```

```
user =[]
for i in range(len(pre.columns)):
  user.append(pre.columns[i])
#print(user)

for i in range(len(user)):
  user[i] = user[i] + "_pre"
#print(user)

pre.columns = user
pre.head()
```

```
user =[]
for i in range(len(post.columns)):
  user.append(post.columns[i])
#print(user)

for i in range(len(user)):
  user[i] = user[i] + "_post"
#print(user)

post.columns = user
post.head()
```

```
df_res = pd.concat([pre, post, cg], axis=1)
```

```
df_res = df_res.reindex(sorted(df_res.columns), axis=1)
```
